```python
books = books[["ISBN", "Book-Title", "Book-Author", "Year-Of-Publication", "Publisher", "Image-URL-L"]]
```

```python
books.rename(columns={
    "Book-Title": "Title",
    "Book-Author": "Author",
    "Year-Of-Publication": "Year",
    "Publisher": "Publisher",
    "Image-URL-L": "ImageURL"
}, inplace=True)
```

```python
users = pd.read_csv("./data/Users.csv")
users.shape
```

```python
users.columns
```

```python
users.rename(columns={
    "User-ID": "UserID"
}, inplace=True)
```

```python
users.head()
```

```python
ratings = pd.read_csv("./data/Ratings.csv")
ratings.shape
```

```python
ratings.head()
```

```python
ratings.rename(columns=
               {"User-ID": "UserID",
               "Book-Rating": "BookRating"}, inplace=True)
```

```python
ratings.head()
```

```python
ratings["UserID"].value_counts()
```

```python
x = ratings["UserID"].value_counts() > 50
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
ratings = ratings[ratings["UserID"].isin(y)]
```

```python
ratings.shape
```

```python
books_with_ratings = books.merge(ratings, on="ISBN")
```

```python
books_with_ratings.head(3)
```

```python
number_ratings = books_with_ratings.groupby('Title')['BookRating'].count().reset_index()
```

```python
number_ratings.head()
```

```python
number_ratings.rename(columns={
    "BookRating": "TotalRatingCount"
},inplace=True)
```

```python
final_rating = books_with_ratings.merge(number_ratings,on='Title')
```

```python
final_rating = final_rating[final_rating["TotalRatingCount"] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(["UserID", "Title"],inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
book_pivot = final_rating.pivot_table(index='Title', columns='UserID', values='BookRating', fill_value=0)
 
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model =  NearestNeighbors(algorithm='brute')#, metric='cosine')
model.fit(book_sparse)
```

```python
# lets extract a book name
book_sample = book_pivot.loc["Harry Potter and the Chamber of Secrets (Book 2)",:].values
book_sample
```

```python
distance,suggestion = model.kneighbors(book_sample.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
book_names = book_pivot.index
```

```python
for ind in suggestion[0]:
    print(book_pivot.index[ind])
    # print image url
    print(final_rating[final_rating["Title"] == book_names[ind]]["ImageURL"].values[0])
```

```python
import os, sys, shutil

def use_without_training():
    try:
        os.makedirs("artifacts", exist_ok=True)

        for item in os.listdir("backup"):
            src = os.path.join("backup", item)
            dst = os.path.join("artifacts", item)

            if os.path.isdir(src):
                # copy folder recursively
                if os.path.exists(dst):
                    shutil.rmtree(dst)
                shutil.copytree(src, dst)
            else:
                # copy single file
                shutil.copy2(src, dst)

        print("Using existing model from backup folder")

    except Exception as e:
        print(f"Error using existing model: {e}")
        sys.exit(1)
```

```python
use_without_training()
```
